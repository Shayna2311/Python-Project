```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
sector = ['Power', 'Industry', 'Transport', 'Others']
contribution = [39.2, 45.1, 13.8, 1.8]
plt.title("Emissions Profile 2019")
colors = ('grey','turquoise','beige','wheat')
plt.pie(contribution, colors = colors, labels = sector, autopct= "%0.1f%%")
```

```python
df = pd.read_csv("air_quality_targets_2020.csv")
fig,ax = plt.subplots(figsize=(9,9))
plt.scatter(df.Air_Pollutant, df.Target, label='Target')
plt.scatter(df.Air_Pollutant, df.Actual,label='Actual')
plt.title("Comparison of air quality targets and actual air quality in 2020")
plt.ylabel("Quantity in microgram per cubic metre")
plt.xlabel("Air Pollutants")
plt.legend()
```

```python
df1 = pd.read_csv("air_pollutant.csv")
plt.plot(df1.Year,df1.Sulphur_Dioxide,label='Sulphur Dioxide')
plt.plot(df1.Year,df1.Nitrogen_Dioxide,label='Nitrogen Dioxide')
plt.plot(df1.Year,df1.Particulate_Matter,label='Particulate Matter')
plt.plot(df1.Year,df1.Ozone,label='Ozone')
plt.title("Trends in quantities of different air pollutants over the last 10 years")
plt.xlabel('Year')
plt.ylabel('Quantity in microgram per cubic metre')
plt.legend()
```

```python
plt.plot(df1.Year,df1.Carbon_Monoxide,label='Carbon Monoxide')
plt.title("Trends in quantity of carbon monoxide over the last 10 years")
plt.xlabel('Year')
plt.ylabel('Quantity in microgram per cubic metre')
plt.legend()
```
